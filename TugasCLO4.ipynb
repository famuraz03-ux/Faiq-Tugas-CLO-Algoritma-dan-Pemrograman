Pengurutan (Sort) "Bubble Sort"

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    
data = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(data)
print("Hasil Bubble Sort: ", data)
```

Pengurutan (Sort) "Selection Sort"

```python
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        
data = [64, 34, 25, 12, 22, 11, 90]
selection_sort(data)
print("Hasil Selection Sort: ", data)
```

Pengurutan (Sort) "Insertion Sort"

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        
data = [64, 34, 25, 12, 22, 11, 90]
insertion_sort(data)
print("Hasil Insertion Sort: ", data)
```

Pengurutan (Sort) "Merge Sort"

```python
def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        l = arr[:mid]
        r = arr[mid:]
        
        merge_sort(l)
        merge_sort(r)
        
        i = j = k = 0
        while i < len(l) and j < len(r):
            if l[i] < r[j]:
                arr[k] = l[i]
                i += 1
            else:
                arr[k] = r[j]
                j += 1
            k += 1
            
        while i < len(l):
            arr[k] = l[i]
            i += 1
            k += 1
        while j < len(r):
            arr[k] = r[j]
            j += 1
            k += 1
            
data = [64, 34, 25, 12, 22, 11, 90]
merge_sort(data)
print("Hasil Merge Sort: ", data)
```

Pengurutan (Sort) "Quick Sort"

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    
    pivot = arr[0]
    left = [x for x in arr[1:] if x < pivot]
    right = [x for x in arr[1:] if x >= pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

data = [64, 34, 25, 12, 22, 11, 90]
sorted_data = quick_sort(data)
print("Hasil Quick Sort: ", sorted_data)
```

Pencarian (Search) "Sequential Search"

```python
def sequential_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i # Mengembalikan indeks
    return -1 # Data tidak ditemukan

# Contoh Penggunaan
data_fisika = [10, 50, 30, 70, 80, 20]
target = 50

hasil = sequential_search(data_fisika, target)

if hasil != -1:
    print(f"Data ditemukan pada indeks: {hasil}")
else:
    print("Data tidak ditemukan.")
```

Pencarian (Search) "Binary Search"

```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = low + (high - low) // 2
        
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
            
    return -1

# Data HARUS terurut
data_energi = [2, 5, 8, 12, 16, 23, 38, 56, 72, 91]
target = 72

hasil = binary_search(data_energi, target)

if hasil != -1:
    print(f"Energi ditemukan pada indeks: {hasil}")
else:
    print("Energi tidak ditemukan.")
```

Pencarian dalam Tabel Massa Jenis

```python
class element:
    def __init__(self, z, density, name):
        self.z = z
        self.density = density
        self.name = name
        
def find_element_binary(table, target_z):
    low = 0
    high = len(table) - 1
    
    while low <= high:
        mid = low + (high - low) // 2
        current_element = table[mid]
        
        if current_element.z == target_z:
            return mid
        elif current_element.z < target_z:
            low = mid + 1
        else:
            high = mid - 1
            
    return -1
# Dataset (List of Objective)
periodic_table = [
    element(1, 0.000089, "Hydrogen"),
    element(2, 0.000178, "Helium"),
    element(6, 2.26, "Carbon"),
    element(13, 2.70, "Aluminum"),
    element(14, 2.33, "Silicon"),
    element(26, 7.87, "Iron"),
    element(29, 8.96, "Copper"),
    element(79, 19.32, "Gold"),
]

search_z = 29 # Mencari Copper (Tembaga)
index = find_element_binary(periodic_table, search_z)

print("--- Hasil Pencarian ---")
if index != -1:
    el = periodic_table[index]
    print(f"Unsur: {el.name}")
    print(f"Nomor Atom: {el.z}")
    print(f"Massa Jenis: {el.density} g/cm^3")
else:
    print(f"Unsur dengan Z = {search_z} tidak ditemukan.")
```
